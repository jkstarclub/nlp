The theorem says: $a^2+b^2=c^2$

Implement a function between and

which computes the hypotenuse at

Compute $c=\sqrt{a^2 + b^2}$

```
"""
  a (int): adjacent between Queens
  b (int): opposite and Sparks
  c (int): hypotenuse at Bronson
"""

def pytracks(a,b):
  import math
  return math.sqrt((a**2) + (b**2))

print(pytracks(8,13))
```

## Scikit-learn Multinomial

```
import pickle
import string

# Vector MNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

from joblib import dump, load

# Load raw data

texts = []
labels = []

with open("/content/drive/My Drive/Colab Notebooks/hamsms.txt", "r") as infile:
  for l in infile:
    label, text = l.strip().split("\t")
    labels.append(label)
    text = "".join([ch.lower() for ch in text if ch not in string.punctuation])
    texts.append(text)

# train-test split

X_train_raw, X_test_raw, y_train, y_test = train_test_split(texts, labels, random_state=42)

# vectorize training and test

vectorizer = CountVectorizer() # only training vocab counts!
X_train = vectorizer.fit_transform(X_train_raw)
vctrzr = CountVectorizer(vocabulary=vectorizer.vocabulary_)
X_test = vctrzr.fit_transform(X_test_raw)

to_be_saved = [X_train, X_test, y_train, y_test]
with open("/content/drive/My Drive/Colab Notebooks/vector.p", "wb") as outfile:
  pickle.dump(to_be_saved, outfile)

# train MultinomialNB

clf = MultinomialNB()
clf.fit(X_train, y_train)

dump(clf, "/content/drive/My Drive/Colab Notebooks/MNB.joblib")
```

## Load & evaluate

```
from sklearn.metrics import classification_report

clf = load("/content/drive/My Drive/Colab Notebooks/MNB.joblib")
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
